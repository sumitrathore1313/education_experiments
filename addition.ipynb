```python
import numpy as np
import pandas as pd
```

```python
num1 = np.random.randint(0, 1000, size=10000)
num2 = np.random.randint(0, 1000, size=10000)
ans = num1+num2
```

```python
temp = np.concatenate((num1.reshape(num1.shape[0], 1), num2.reshape(num2.shape[0], 1), ans.reshape(ans.shape[0], 1)), axis=1)
data_frame = pd.DataFrame(temp, columns=['num1', 'num2', 'ans'])
data_frame.to_csv('data/addition.csv')
```

```python
temp = []
for num in num1:
    bin_num = bin(num)[2:]
    temp1 = [0 for i in range(12)]
    for i in range(len(bin_num)):
        temp1[-i-1] = int(bin_num[-i-1])
    temp.append(temp1)
bin_num1 = temp
temp = []
for num in num2:
    bin_num = bin(num)[2:]
    temp1 = [0 for i in range(12)]
    for i in range(len(bin_num)):
        temp1[-i-1] = int(bin_num[-i-1])
    temp.append(temp1)
bin_num2 = temp
temp = []
for num in ans:
    bin_num = bin(num)[2:]
    temp1 = [0 for i in range(12)]
    for i in range(len(bin_num)):
        temp1[-i-1] = int(bin_num[-i-1])
    temp.append(temp1)
train_y = temp
train_x =  np.concatenate((bin_num1, bin_num2), axis=1)
print(train_x[31])
print(bin(num1[31]))
print(bin(num2[31]))
print(bin(ans[31]))
                             
```

```python
from keras.models import Sequential

model = Sequential()

from keras.layers import Dense
model.add(Dense(units=64, activation='relu', input_dim=24))
model.add(Dense(units=128, activation='relu'))
model.add(Dense(units=12, activation='sigmoid'))

model.compile(loss='categorical_crossentropy',
              optimizer='sgd',
              metrics=['accuracy'])
# x_train and y_train are Numpy arrays --just like in the Scikit-Learn API.
model.fit(train_x, train_y, epochs=100, batch_size=10)
```

```python
y_pred = model.predict(train_x)
print(y_pred[0:10])
print(train_y[0:10])
```
